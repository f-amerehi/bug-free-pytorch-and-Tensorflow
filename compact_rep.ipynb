```
import requests
import blurhash
import numpy as np
from PIL import Image
from thumbhash import image_to_thumbhash
from thumbhash import thumbhash_to_image
```

```

#url = 'https://t3.ftcdn.net/jpg/05/57/73/24/360_F_557732439_jQDlg9VwJDR7sZOj2urhhAZhAJ03oJVR.jpg'
#url = 'https://i.pinimg.com/1200x/6a/48/be/6a48be32f5dbd8611971f1153b2f38dc.jpg'
url = 'https://img-aws.ehowcdn.com/360x267p/s3-us-west-1.amazonaws.com/contentlab.studiod/getty/060822b55d424d86b43c5abd1a2e0fd4'

img = Image.open(requests.get(url, stream=True).raw)
H, W = 200, 300
Resize_image = img.resize((W, H ))
Resize_image
```

### Blurhash encode decode

```
blurhashed = blurhash.encode(np.array(Resize_image))
print(blurhashed)
```

```
Image.fromarray(np.array(blurhash.decode(blurhashed, W, H)).astype('uint8'))
```
